```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

```

```python
data_folder = 'D:\pycharm\jupyter'
file_names =['midu113318.xlsx','midu113320.xlsx','midu113321.xlsx','midu113322.xlsx','midu113323.xlsx','midu113324.xlsx','midu113325.xlsx','midu113326.xlsx','midu113327.xlsx','midu113328.xlsx','midu113330.xlsx','midu113331.xlsx',
            'midu113332.xlsx','midu113333.xlsx','midu113334.xlsx','midu113335.xlsx','midu113337.xlsx','midu113338.xlsx','midu113339.xlsx','midu113341.xlsx',]
```

```python
data_list = []

train_data_list = []
validation_data = None
scaler = MinMaxScaler()  # 创建一个MinMaxScaler对象
for i, file_name in enumerate(file_names):
    file_path = os.path.join(data_folder, file_name)
    data = pd.read_excel(file_path, engine='openpyxl')
    features = data.iloc[500:, 1:31].values  # 剔除前 500 个时序点，选择第2到第31列作为特征
    labels = data.iloc[500:, 2].values  # 原文件的第三列作为标签数据
    
    # 最大最小值归一化处理
    features_scaled = scaler.fit_transform(features)  # 使用同一个scaler对象对特征数据进行归一化处理
    labels_scaled = scaler.fit_transform(labels.reshape(-1, 1)).flatten()  # 使用同一个scaler对象对标签数据进行归一化处理
    
    if i < 19:
        train_data_list.append((features_scaled, labels_scaled))
    else:
        validation_data = (features_scaled, labels_scaled)
```

```python
# 创建训练数据和标签数据的列表
X_train_list = []
y_train_list = []

for features, labels in train_data_list:
    num_sequences = len(features) - 60 + 1
    for i in range(num_sequences):
        X_train_list.append(features[i:i+50, :])  # 将每天的数据分割成长度为50的序列
        y_train_list.append(labels[i+50:i+60])  # 预测的是后10个时序点的第三列数据（标签）
```

```python
# 将列表转换为 NumPy 数组
X_train = np.array(X_train_list)
y_train = np.array(y_train_list)

# 构建 LSTM 模型
model = Sequential()
model.add(LSTM(64, input_shape=(50, 30)))  # 输入特征维度为(50, 30)
model.add(Dense(10))

# 编译模型
model.compile(loss='mse', optimizer='adam')
```

```python
# 训练模型
history = model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1, validation_data=validation_data)

```
