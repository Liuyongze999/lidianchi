```python
import pandas as pd
from sklearn.linear_model import LinearRegression

import numpy as np


# 命名规范
v_columns = ["V"+str(i) for i in range(1,18)]
TSU_columns = ["TSU"+str(i) for i in range(1,18)]
TSD_columns = ["TSD"+str(i) for i in range(1,18)]
TU_columns = ["TU"+str(i) for i in range(1,18)]
TD_columns = ["TD"+str(i) for i in range(1,18)]

data_path = './锂离子电池生产参数调控及生产温度预测挑战赛公开数据-初赛/'
traindata = pd.read_csv(data_path + 'train.csv')
testdata = pd.read_csv(data_path + 'test.csv')
submit = pd.read_csv('./提交示例.csv')
submit.head()
traindata=traindata.drop(1057)
```

```python
traindata.columns = ["Num","Time"]\
+v_columns\
+TSU_columns\
+TSD_columns\
+TU_columns\
+TD_columns

testdata.columns = ["Num","Time"]\
+v_columns\
+TSU_columns\
+TSD_columns\

submit.columns = ["Num"]\
+TU_columns\
+TD_columns

# traindata.drop(["Num","Time"],axis=1,inplace=True)
# testdata.drop(["Num","Time"],axis=1,inplace=True)
traindata.head()
```

```python
label = ["TU"+str(i) for i in range(1,18)]\
+["TD"+str(i) for i in range(1,18)]

features = ["V"+str(i) for i in range(1,18)]\
+["TSU"+str(i) for i in range(1,18)]\
+["TSD"+str(i) for i in range(1,18)]
## 构造V 流量特征  VB
## V和VB1  V均值VB2
traindata["VB1"] = traindata[v_columns].apply(lambda x:x.sum(),axis=1)
traindata["VB2"] = traindata[v_columns].apply(lambda x:x.sum()/17,axis=1)
testdata["VB1"] = testdata[v_columns].apply(lambda x:x.sum(),axis=1)
testdata["VB2"] = testdata[v_columns].apply(lambda x:x.sum()/17,axis=1)
```

```python
## V上面7个和  均值
traindata["VB3"] = traindata[v_columns[:7]].apply(lambda x:x.sum(),axis=1)
traindata["VB4"] = traindata[v_columns[:7]].apply(lambda x:x.sum()/7,axis=1)
testdata["VB3"] = testdata[v_columns[:7]].apply(lambda x:x.sum(),axis=1)
testdata["VB4"] = testdata[v_columns[:7]].apply(lambda x:x.sum()/7,axis=1)
## 利用时间 构造流量速率特征  对于我们只预测TU1-7那就再只做一下V1-7的流速
## 流速VS1-17 命名  注意这里还没有取绝对值  也就是说 这里的流速有正负值  同理17个流速都可以获得
## 这里直接做VS1-17的流速
traindata["Time1"] = pd.to_datetime(traindata["Time"]) - pd.to_datetime(traindata["Time"].shift(periods=1))
testdata["Time1"] = pd.to_datetime(testdata["Time"]) - pd.to_datetime(testdata["Time"].shift(periods=1))
# 筛选时差20分钟内  且不为0 的index
indexset = traindata[(traindata["Time1"] <= pd.Timedelta(days=0, minutes=20)) & (traindata["Time1"] != pd.Timedelta(days=0, minutes=0))].index
```

```python
for i in range(1,18):
    traindata["VS"+str(i)] = np.nan
    traindata.loc[indexset,"VS"+str(i)] = (traindata.loc[indexset]["V"+str(i)] - traindata.loc[indexset]["V"+str(i)].shift(periods=1)) / traindata.loc[indexset]["Time1"].apply(lambda x:x.seconds//60)
    traindata.loc[indexset,"VS"+str(i)]
    # 后值填充  又做了个取绝对值
    traindata["VS"+str(i)] = traindata["VS"+str(i)].fillna(method="bfill").apply(lambda x:abs(x))
# 筛选时差20分钟内  且不为0 的index
indexset = testdata[(traindata["Time1"] <= pd.Timedelta(days=0, minutes=20)) & (testdata["Time1"] != pd.Timedelta(days=0, minutes=0))].index
for i in range(1,18):
    testdata["VS"+str(i)] = np.nan
    testdata.loc[indexset,"VS"+str(i)] = (testdata.loc[indexset]["V"+str(i)] - testdata.loc[indexset]["V"+str(i)].shift(periods=1)) / testdata.loc[indexset]["Time1"].apply(lambda x:x.seconds//60)
    testdata.loc[indexset,"VS"+str(i)]
    # 后值填充  又做了个取绝对值
    testdata["VS"+str(i)] = testdata["VS"+str(i)].fillna(method="bfill").apply(lambda x:abs(x))
## 加热棒设定温度和流量的比值
## TSUV  TSDV
for i in range(1,18):
    traindata["TSUV"+str(i)] = traindata["TSU"+str(i)] / traindata["V"+str(i)]
    traindata["TSDV"+str(i)] = traindata["TSD"+str(i)] / traindata["V"+str(i)]
    ## 加热棒设定温度和流量的比值
## TSUV  TSDV
for i in range(1,18):
    testdata["TSUV"+str(i)] = testdata["TSU"+str(i)] / testdata["V"+str(i)]
    testdata["TSDV"+str(i)] = testdata["TSD"+str(i)] / testdata["V"+str(i)]

##判断是否工作日  没有用
traindata["day"] = pd.to_datetime(traindata["Time"]).apply(lambda x:x.dayofweek)
testdata["day"] = pd.to_datetime(testdata["Time"]).apply(lambda x:x.dayofweek)
```

```python
import lightgbm as lgb
traindata = traindata.fillna(method='bfill', axis=0)
testdata = testdata.fillna(method='bfill', axis=0)
for col in ['TU1', 'TU2', 'TU3', 'TU4', 'TU5', 'TU6', 'TU7', 'TD1', 'TD2', 'TD5']:
    # 创建LightGBM回归模型并进行拟合
    model = lgb.LGBMRegressor()
    model.fit(traindata.drop(['Time1', 'Time','TU1', 'TU2', 'TU3', 'TU4', 'TU5', 'TU6', 'TU7', 'TU8', 'TU9', 'TU10', 'TU11', 'TU12', 'TU13', 'TU14', 'TU15', 'TU16', 'TU17', 'TD1', 'TD2', 'TD3', 'TD4', 'TD5', 'TD6', 'TD7', 'TD8', 'TD9', 'TD10', 'TD11', 'TD12', 'TD13', 'TD14', 'TD15', 'TD16', 'TD17'], axis=1), traindata[col])
    
    # 预测测试数据的值
    submit[col] = model.predict(testdata.drop(['Time','Time1'], axis=1))
```
